# Step 1 - Loading the Image Data

```python
import PIL
import numpy as np

from PIL import Image
from numpy import asarray

example_image = PIL.Image.open("blue-5.png")
example_image
```

```python
asarray(example_image)
```

```python
# Load all the images in for our dataset....

images = []
targets = []
target_names = ['Black', 'Blue', 'Green', 'Orange', 'Pink', 'Red', 'White', 'Yellow']

for target in target_names:
    count = 1
    while True:
        try:
            image_name = "{}-{}.png".format(target.lower(), count)
            print("Loading image: " + image_name)
            images.append(asarray(PIL.Image.open(image_name)))
            targets.append(target_names.index(target))
            count = count + 1
        except:
            break        
```

# Step 2 - Processing the data
We need to decide which features from the data  
will be useful for our machine learning algorithm.

## Idea 1
We could calculate the mean of the Red, Green and Blue for every
pixel in the dataset.  Let's try that with a single image first.

```python
images[10]
```

Explore the shape to understand the structure of the image.

```python
images[10].shape
```

So we can get the Red value by doing:

```python
red = images[10][:,:,0]
print(red)
```

And the green and blue by doing...

```python
green = images[10][:,:,1]
blue = images[10][:,:,2]
print(green)
print(blue)
```

So, we can get the mean via...

```python
R = np.mean(images[10][:,:,0])
G = np.mean(images[10][:,:,1])
B = np.mean(images[10][:,:,2])
print (R, G, B)
```

Right, let's put that in a re-usable function called
transform...

```python
def transform(the_image):
    
    R = np.mean(the_image[:,:,0])
    G = np.mean(the_image[:,:,1])
    B = np.mean(the_image[:,:,2])
    
    return np.array([R, G, B])
```

And try it out...

```python
transform(images[10])
```

```python
transform(images[5])
```

Now, let's transform all our images... and build a new dataset.
We could do this one image at a time.

```python
transformed_images = []

for i in images:
    transformed_images.append ( transform(i) )

transformed_images
```

This is a little slow, we have a quicker way, using something 
known as a for comprehension.

```python
transformed_images = [ transform(i) for i in images ]
```

```python
transformed_images
```

# Step 3 - Split and Train
Next, we need to split our data into a training set, and a model set.  We can use the normal train_test_split function from scikit learn.

```python
from sklearn.model_selection import train_test_split

X_data, Y_data, X_target, Y_target = train_test_split(
    transformed_images,
    targets,
    test_size=0.33
)
```

Note the convention.... X is usually used to represent the training set, and Y the testing set.

```python
from sklearn import svm
classifier = svm.SVC(kernel='linear')
classifier.fit(X_data, X_target)
```

# Step 4 - Test our model...
Now we can test the model, to see how it performs.

```python
result = classifier.predict(Y_data)
result
```

```python
Y_target
```

```python
np.mean(result == Y_target) * 100
```

# Step 5 - Can we improve the accuracy?
84% isn't a bad initial go... but, we can do better. 
First, let's try the other models (gauisan, poly, etc.)

```python
classifier = svm.SVC(kernel='poly')
classifier.fit(X_data, X_target)
result = classifier.predict(Y_data)
np.mean(result == Y_target) * 100
```

They do okay, but not the best. So, what else could we change?

The features.  When we took the mean, we effectively lost information, we reduced our feature set.  Let's expand it.

Let's try feeding every single pixel into the processor (by the way, more features isn't always best, and it will slow the process down).

We need a new transform function.... fortunately, given we have a numpy array for our images, we can just adjust the shape.

```python
images[10].shape
```

```python
print(images[10])
```

```python
re_shaped = images[10].reshape(1,30000)
re_shaped
```

The problem is, we end up with a two-dimensional array...

```python
re_shaped.ndim
```

Numpy has a neat feature - squeeze...

```python
squeezed_and_reshaped = np.squeeze(re_shaped)
squeezed_and_reshaped
```

```python
squeezed_and_reshaped.ndim
```

Perfect.. let's put it in a new transform function

```python
def transform(the_image):
    
    re_shape = the_image.reshape(1, 30000)
    return np.squeeze(re_shape)
```

Try it with a few images...

```python
transform(images[10])
```

```python
transform(images[5])
```

Right, as before - process our images...

```python
transformed_v2_images = [ transform(i) for i in images ]
```

```python
transformed_v2_images
```

We're seeing pretty similar results.  Why?

Well, here's the interesting thing.  Even though we /said/ we lost information, all our images were flat - so, the mean of the same value over and over again has no effect.

In other words, our "compressed features", didn't actually lose anything. So, what else could be going on?

```python
new_image_list = []
new_target_list = []

def expand_data_set(image_list, target_list, image, target):
    
    image_plus_five_pct = image[:,:,:] * 1.05
    image_minus_five_pct = image[:,:,:] * 0.95
    
    image_list.append(image) # add the original
    image_list.append(image_plus_five_pct) # the plus 5%
    image_list.append(image_plus_five_pct) # the minus 5%
    
    target_list.append(target)
    target_list.append(target)
    target_list.append(target)
    
for i, t in zip(images,targets) :
    
    expand_data_set(new_image_list,
                    new_target_list,
                    i,
                    t)
```

# Going further...
Okay, that's fairly simple when we just have a static image.
But, what about when we have more complex images - a cat, a mouse, etc. How do we find and classify those things?

Well, the general principle is the same -

 1. Load the data
 2. Process the images to be consistent in format
 3. Identify key features for the learning algorithm
 4. Build a model
 5. Test the model
 
The hard part is 3 - identifying the features.

A good starting point for 'is this an image of...' questions is HOG - it stands for Histogram of Oriented Gradiends.  Basically, it takes a set of pixels from your images, say 8 x 8, and calculates for that set of pixels the change in values. This then becomes a feature in the learner, and then we take the next 8x8 grid and do the same again...

We could build this from scratch... but, scikit learn has a sub-package called skimage....

```python
from skimage.feature import hog
```

And now we can use this to convert an image into a HOG...

```python
blue_image = asarray(PIL.Image.open("blue-5.png"))

blue_hog, blue_hog_img = hog(
    blue_image,
    pixels_per_cell=(8,8),
    visualize=True,
)
```

```python
# what features do we have
blue_hog
```

```python
# what does the 'HOG' versions of the image look like?
blue_hog_img 
```

Why is it all zeros? Because there's no *change* from one pixel to another in our plain blue image...

Let's try on a picture of a bird...

```python
bird_image = asarray(PIL.Image.open("bird.png"))

bird_hog, bird_hog_img = hog(
    bird_image,
    pixels_per_cell=(8,8),
    visualize=True,
)
```

```python
bird_hog
```

We can visualize this using matplotlib.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2)
fig.set_size_inches(8, 6)
ax[0].imshow(bird_image)
ax[0].set_title("Bird")
ax[1].imshow(bird_hog_img)
ax[1].set_title("Hog")
plt.show()
```

Notice the changes are more prevelant where the bird is, because the gradient change on the background isn't significant.

So, we can now build a training set from the hog values, and see if we can classify images of animals - or hand writing, or other things! 

As a final couple of notes:

 1. Our feature list should always be the same size. We had no issues, our images were all the same size, so our feature sets always came out the same size.
       
    But, this is something you might need to work on.
    
       
 2. We *may* get better results by converting our images to grayscale - if colour isn't of interest.
       
    It was of interest for our first example, but not for 'detecting' a bird...
    
For both of these kind of things, the skimage package can help us.

```python
from skimage.io import imread
from skimage.transform import rescale

bird_image = imread('bird.png', as_gray=True)
bird_image = rescale(bird_image, 1/3, mode='reflect')

bird_hog, bird_hog_img = hog(
    bird_image,
    pixels_per_cell=(8,8),
    visualize=True,
)
```

```python
bird_hog
```

```python
fig, ax = plt.subplots(1,2)
fig.set_size_inches(8, 6)
ax[0].imshow(bird_image, cmap='gray')
ax[0].set_title("Bird")
ax[1].imshow(bird_hog_img, cmap='gray')
ax[1].set_title("Hog")
plt.show()
```
